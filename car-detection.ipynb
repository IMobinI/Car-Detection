# *Train Model:*

```python
# Initialize model
model = YOLO('yolov8n.pt')

# Train the model
results = model.train(data='/kaggle/input/car-images/Dataset/data.yaml', epochs=10, imgsz=640)
```

# *Evaluating Accuracy on the Test Set:*

```python
results = model.val(data='/kaggle/input/car-images/Dataset/data.yaml', split='test')
```

### show sample:

```python
# List of image paths
image_paths = [
    '/kaggle/input/car-images/Dataset/test/images/onewayup_mp4-0008_jpg.rf.4632b2082e9ed9ef4379284717ef7e4c.jpg',
    '/kaggle/input/car-images/Dataset/test/images/onewayup_mp4-0022_jpg.rf.f84a7544129d7990445c902a06020e2a.jpg',
    '/kaggle/input/car-images/Dataset/test/images/onewayup_mp4-0026_jpg.rf.08d9942d361b3e06ac854f4c59c5f969.jpg'
]

num_images = len(image_paths)

fig, axes = plt.subplots(1, num_images, figsize=(15, 15))

for idx, image_path in enumerate(image_paths):
    # Perform inference
    results = model.predict(source=image_path)

    # Extract the first result
    result = results[0]

    # Get the image with annotations
    image_with_annotations = result.plot()

    # Convert the image to RGB
    image_with_annotations = cv2.cvtColor(image_with_annotations, cv2.COLOR_BGR2RGB)

    # Display the image with predictions
    if num_images == 1:
        axes.imshow(image_with_annotations)
        axes.axis('off')
    else:
        axes[idx].imshow(image_with_annotations)
        axes[idx].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
