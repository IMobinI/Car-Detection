# *Train Model:*

```python
# Initialize model
model = YOLO('yolov8n.pt')

# Train the model
results = model.train(data='/kaggle/input/car-images/Dataset/data.yaml', epochs=50, imgsz=640)
```

# *Evaluating Accuracy on the Test Set:*

```python
results = model.val(data='/kaggle/input/car-images/Dataset/data.yaml', split='test')
```
